```python
import os
import shutil

parent_dir = '.'

for item in os.listdir(parent_dir):
    # Construct the path to the current item
    item_path = os.path.join(parent_dir, item)
    
    # Check if the current item is a directory and matches the NC_XXXX pattern
    if os.path.isdir(item_path) and item.startswith('NC_'):
        # Define paths to the 'all_islands' and 'tfidf' directories
        all_islands_path = os.path.join(item_path, 'patterns', 'all_islands')
        tfidf_path = os.path.join(item_path, 'patterns', 'tfidf')
        
        # Check if both 'all_islands' and 'tfidf' directories exist
        if os.path.isdir(all_islands_path) and os.path.isdir(tfidf_path):
            # Define the new destination path for 'all_islands'
            new_all_islands_path = os.path.join(tfidf_path, 'all_islands')
            
            # Move 'all_islands' into 'tfidf'
            shutil.move(all_islands_path, new_all_islands_path)
            
            print(f"Moved 'all_islands' to 'tfidf' in {item}")
        else:
            print(f"'all_islands' or 'tfidf' folder not found in {item}")
```
